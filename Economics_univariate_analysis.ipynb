#PLOTS


```
import matplotlib.pyplot as plt
plt.figure(figsize=(20,5))
plt.plot(clean[0]['date'], clean[0]['mean'])
```

```
avg = []
for city in clean:
    avg.append(city.groupby('date').agg({'mean': ['mean', 'min', 'max']}).reset_index())
```

```
avg
```

```
city_names = [name[:name.index('_')] for name in file_names]
for i in range(len(avg)):
    city = avg[i]
    city.plot(x='date', y='mean', figsize=(20, 10), kind='line', title=city_names[i])
```

```
city_names
```

```
change = []
for city in avg:
    init = float(city.loc[0]['mean']['mean'])
    final = float(city.loc[117]['mean']['mean'])
    change.append(final - init)
change
```

```
plt.figure(figsize=(15,5))
plt.bar(city_names, change)
plt.title("Average drop in each city plot")
```
